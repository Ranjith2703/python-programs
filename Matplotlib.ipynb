```python
years = ["2019", "2020", "2021", "2022"]
cse = [150, 175, 200, 275]
ece = [125, 150, 150, 250]
eee = [100, 125, 125, 225]
eee_start = [cse[i] + ece[i] for i in range(len(cse)) ]
plt.title ("PLACEMENT COMPARISON")
plt.barh (years, cse, height = 0.4, color = 'Red', label = 'cse')
plt.barh (years, ece, left  = cse,  height = 0.4, color = 'Green', label = 'ece')
plt.barh (years, eee, left = eee_start, height = 0.4, color = 'Blue', label = 'eee')
plt.xlim(0,700)
plt.xlabel('Placements')
plt.ylabel('Years')
plt.legend()
plt.show()
```

# GROUPED BAR CHART|| MULTIPLE BAR CHART EXAMPLE

```python
years = ["2019", "2020", "2021", "2022"]
cse = [150, 175, 200, 275]
ece = [125, 150, 150, 250]
eee = [100, 125, 125, 225]
w = 0.2

plt.figure(figsize=(10,5))

cse_bar = np.arange(0, len(cse))
print(cse_bar)

ece_bar = [i + w for i in cse_bar]
print(ece_bar)

eee_bar = [i + w for i in ece_bar]
print(eee_bar)

plt.bar (cse_bar, cse, label = 'cse', width = w, color = 'Red')
plt.bar (ece_bar, ece, label = 'ece', width = w, color = 'Green')
plt.bar (eee_bar, eee, label = 'eee', width = w, color = 'Blue')

plt.xticks(cse_bar + w + w/2, years)

plt.title ("PLACEMENT COMPARISON")
plt.xlabel("YEARS")
plt.ylabel("Placement Count")
plt.legend()
plt.show()
```

# PIE CHART - labels, startangle, explode, shadow, colors, legend, autopct

# HISTOGRAM CHART

```python
marks = [95, 85, 65, 55, 54, 92, 88, 96]
grade_intervals = [0, 50, 60, 70, 80 ,90, 100]
plt.title("Student's Grade")
plt.xlim(0,100)
plt.ylim(0,100)
plt.hist(marks, grade_intervals)
plt.xticks([0, 50, 60, 70, 80 ,90, 100])
plt.xlabel("Marks obtained")
plt.show()
```

```python
plt.hist(marks, grade_intervals, histtype ='bar', rwidth = 0.7)
plt.xticks([0, 50, 60, 70, 80 ,90, 100])
plt.xlabel("Marks obtained")
plt.show()
```

```python
plt.hist(marks, grade_intervals, histtype ='stepfilled', rwidth = 0.7, facecolor ='green')
plt.xticks([0, 50, 60, 70, 80 ,90, 100])
plt.xlabel("Marks obtained")
plt.show()
```

# VISUALIZATION WITH ERROR

# MESH GRID

```python
x = np.linspace(-4, 4, 9)

# numpy.linspace creates an array of
# 9 linearly placed elements between
# -4 and 4, both inclusive
y = np.linspace(-5, 5, 11)
print("before x.shape" , x.shape)
print("before y.shape" , y.shape)
# The meshgrid function returns
# two 2-dimensional arrays
x_1, y_1 = np.meshgrid(x, y)
print(x,y,sep="\n")
print("x_1 = ")
print(x_1)
print("y_1 = ")
print(y_1)
print(x_1.shape,y_1.shape,sep="\n")
```
