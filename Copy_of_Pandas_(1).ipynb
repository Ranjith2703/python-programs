# Series
* A pandas Series is a **one-dimensional array**-like object in Python that can hold data of any type (integers, floats, strings, etc.).
* It is part of the pandas library, which is widely used for data manipulation and analysis.

#Key Features of a pandas Series

**One-dimensional:**

* A Series is like a **column** in a table or a single list of values.

**Indexing:**

* Each element in a Series has an associated label called an **index**.
* By default, this index is just the row number (starting from 0), but you can also assign custom labels to it.

**Homogeneous Data Type:**

* All elements in a Series have the same data type, unlike a DataFrame, which can have different types of data in different columns.

**Summary**

* A pandas Series is a powerful tool for handling one-dimensional data with ease.
* It provides flexibility with indexing and supports a wide range of operations, making it an essential component of the pandas library for data analysis in Python.

# DataFrame
* A pandas DataFrame is a two-dimensional, tabular data structure in Python that is part of the pandas library, widely used for data manipulation and analysis.
* Think of it as an Excel spreadsheet or a SQL table in Python.

**Key Features of a pandas DataFrame**

**Two-dimensional:**

* Data is organized in rows and columns, similar to a table.
Heterogeneous Data:

* Different columns can hold different types of data (e.g., integers, floats, strings).

**Labeled Axes:**

* Both rows and columns have labels (indices), which makes it easy to access and manipulate data.

**Summary**

**Series:**
* All elements have the same data type.

**DataFrame:**
* Different columns can have different data types.

     * By keeping data types homogeneous in a Series, pandas ensures efficient storage and computation.

     * This is in contrast to a DataFrame, where the flexibility to store different data types in different columns is useful for handling more complex datasets.

# Sorting
* Sorting is one of the operations performed on the dataframe based on conditional requirements.
* We can sort dataframe alphabetically as well as in numerical order also.

**Method 1: Using sort_values() method**

**Syntax:**
df_name.sort_values(by column_name, axis=0, ascending=True, inplace=False, kind=’quicksort’, na_position=’last’, ignore_index=False, key=None)

# Parameters:
#by:
* name of list or column it should sort by

# axis:
* Axis to be sorted.('0' or 'index' or 1 or ‘column’) by default its 0.(column number)

# ascending:
* Sorting ascending or descending. Specify lists of bool values for multiple sort orders.
* The list of bool values must match the no. of values of ‘by’ i.e. column_names.
* By default it is true.

#inplace:
* By default it is false. but if its value is true it performs operation in-place i.e. in proper place.

#kind:

* Choice of sorting algorithm like quick sort. merge sort, heap sort. by default it is quick sort.

```python
import pandas as pd
technologies = ({
    'Courses':["Spark","Hadoop","pandas","Oracle","Java"],
    'Fee' :[20000,25000,26000,22000,20000],
    'Duration':['30days','35days','40days','50days','60days'],
    'Discount':[1000,2300,1500,1200,2500]
               })
df2 = pd.DataFrame(technologies, index = ['r1','r2','r3','r4','r0'])
df2
```

#Sort Multiple Columns in pandas DataFrame

* By using the sort_values() method you can sort multiple columns in DataFrame by ascending or descending order.
* When not specified order, all columns specified are sorted by ascending order.

```python
# Sort multiple columns
df2.sort_values(['Fee', 'Discount'])
df2
```

```python
#In case if you wanted to update the existing DataFrame use inplace=True.
# Sort ascending order
df2.sort_values(by=['Fee','Discount'], inplace=True)
df2
```

#Sort in an Ascending Order
* Use ascending param to sort the DataFrame in ascending or descending order.
* When you have multiple sorting columns. By default, it sorts in ascending #order.

```python
df2
```

```python
# Sort ascending order
df2.sort_values(by=['Fee','Discount'], inplace=True,
               ascending = [True, True])
df2
```

#Sort Multiple Columns in Descending Order
* In case you wanted to sort by descending order, use ascending=False. You can also specify different sorting orders for each input.

```python
# Sort descending order
df2.sort_values(by=['Fee','Discount'], inplace=True,
               ascending = [True, False])
df2
```

#Group by
* The groupby() method in pandas is used to group data by one or more columns.

#First
* The first() method is often used after groupby() to return the first value in each group.
* This can be particularly useful for getting a representative value from each group.

```python
# importing pandas as pd for using data frame
import pandas as pd

# creating dataframe with student details
dataframe = pd.DataFrame({'id': [7058, 4511, 7014, 7033],
						'name': ['sravan', 'manoj', 'aditya', 'bhanu'],
						'Maths_marks': [99, 97, 88, 90],
						'Chemistry_marks': [89, 99, 99, 90],
						'telugu_marks': [99, 97, 88, 80],
						'hindi_marks': [99, 97, 56, 67],
						'social_marks': [79, 97, 78, 90], })

# group by name
print(dataframe.groupby('name').first())
```

```python
print("------------------------")
# group by name with social_marks sum
print(dataframe.groupby('name')['social_marks'].sum())
```

```python
print("------------------------")
# group by name with maths_marks count
print(dataframe.groupby('name')['Maths_marks'].count())
```
